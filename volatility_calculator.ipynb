```python
import pandas_datareader as pdr
import datetime as dt
import numpy as np

ticker = "GOOGL"
start = dt.datetime(2022, 2, 1)
end = dt.datetime(2022, 4, 22)
data = pdr.get_data_yahoo(ticker, start, end)
data.head()
```

```python
data['Volatility'] = np.log(data['Close']/data['Close'].shift())
data.head()
```

```python
data['Volatility'].std()
data.head()
```

```python
data.to_csv('volatility_googl.csv')
```
