```python
import pandas as pd
import pysentiment2 as ps
```

```python
def polarity_lm(text):
    lm = ps.LM()
    tokens = lm.tokenize(text)
    score = lm.get_score(tokens)
    return score['Polarity']
```

```python
polarity_lm('$AAPL Fed&#39;s Harker sticks with supporting a quarter-point rate hike in March')
```

```python
def polarity_hiv(text):
    hiv4 = ps.HIV4()
    tokens = hiv4.tokenize(text)  
    score = hiv4.get_score(tokens)
    return score['Polarity']
```

```python
polarity_hiv('$AAPL Fed&#39;s Harker sticks with supporting a quarter-point rate hike in March')
```

```python
googl_df = pd.read_csv('posts_googl.csv')
googl_df
```

```python
from tqdm.notebook import tqdm
tqdm.pandas()
polarity_hiv = googl_df['text'].progress_apply(polarity_hiv)
```

```python
googl_df['polarity_hiv4'] = polarity_hiv
```

```python
googl_df
```

```python
polarity_googl = googl_df.loc[googl_df['polarity_hiv4'] != 0]
polarity_googl
```
